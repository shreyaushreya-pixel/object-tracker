```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python

# YOLOv8 Nano Model
model = YOLO("yolov8n.pt")
```

```python

# Replace with your image path
image_path = "car .jpg"

img = cv2.imread(image_path)
```

```python
# Check image loaded or not
if img is None:
    print("Image not found!")
else:
    print("Image Loaded Successfully")
```

```python
results = model(img)
```

```python
annotated_frame = results[0].plot()

# Convert BGR to RGB
annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(annotated_frame)
plt.axis("off")
plt.title("Vehicle Detection Output")
plt.show()
```

```python

output_path = "detected_output.jpg"

cv2.imwrite(output_path,
            cv2.cvtColor(annotated_frame, cv2.COLOR_RGB2BGR))

print("Output Saved Successfully")
```

```python

for r in results:
    boxes = r.boxes

    for box in boxes:
        class_id = int(box.cls[0])
        confidence = float(box.conf[0])

        object_name = model.names[class_id]

        print(f"Detected: {object_name} | Confidence: {confidence:.2f}")
```

```python

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    if not ret:
        break

    # Detect Objects
    results = model(frame)

    # Plot Results
    annotated_frame = results[0].plot()

    cv2.imshow("Vehicle Detection", annotated_frame)

    # Press ESC to Exit
    key = cv2.waitKey(1)

    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
